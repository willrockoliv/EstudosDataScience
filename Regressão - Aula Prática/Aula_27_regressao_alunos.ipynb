# Regressão

```python
# Importação de bibliotecas

import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression

from sklearn.model_selection import train_test_split
```

```python
# Carregamento dos dados

#df = pd.read_csv('idh_dados.csv')

df = pd.read_csv('atlas.csv')
#df = pd.read_csv('atlas2013_dadosbrutos_pt com lat long e altura.csv')
df = df.fillna(df.mean())
```

```python
df
```

```python
quebra = train_test_split(df,train_size=0.8)

df_train = quebra[0]

df_test = quebra[1]
```

```python
# Definindo variável dependente (X) e variável resposta (y)

# O comando reshape(-1, 1) transforma os dados de matriz linha para matriz coluna

X_train = df_train['t_analf18m'].values.reshape(-1, 1)
X_test = df_test['t_analf18m'].values.reshape(-1, 1)


y_train = df_train['idhm'].values.reshape(-1, 1)
y_test = df_test['idhm'].values.reshape(-1, 1) 
```

```python
# Treino do modelo
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
#Avaliação gráfica do ajuste

# Variáveis auxiliares para desenhar o ajuste no gráfico
xfit=np.linspace(0,80,num=1000).reshape(-1,1)
yfit=lr.predict(xfit)
```

```python
# Plotando resultado num gráfico
plt.figure(figsize=(10,7))
plt.scatter(df_test['t_analf18m'],df_test['idhm'], s=20, marker ='o',color='w',edgecolors='black') # Desenha os pontos com os dados
plt.scatter(xfit,yfit,color='r',marker='.',s=10) # Desenha a reta vermelha do ajuste
 
# Opções de formatação
plt.ylabel('IDH',fontsize=20)
plt.xlabel('Taxa de analfabetismo (maiores de 18 anos)',fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=12)
plt.grid()
plt.show()
```

```python
# Analisando resíduos

# Aplicação do modelo aos dados
df_test['idh_predito'] = lr.predict(X_test)
 
# Cálculo dos resíduos
df_test['res'] = df_test['idhm'] - df_test['idh_predito']
```

```python
# Gráfico de distribuição dos resíduos
plt.figure(figsize=(8,4))
plt.hist(df_test['res'])
plt.ylabel('Frequência',fontsize=20)
plt.xlabel('Resíduo',fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=12)
plt.show()
```

```python
# Cálculo do R² do modelo
print("R² =",lr.score(X_test,y_test))
 
# Cálculo do erro quadrático médio do modelo
print("Erro Quadrático Médio =" ,np.mean((y_test - lr.predict(X_test))**2))
```

Escolha do melhor valor de k


```python
#Carregando funções para dividir base em treino e teste

from sklearn.metrics import mean_squared_error

# quebra = train_test_split(df,train_size=0.8)

# df_train = quebra[0]

# df_test = quebra[1]

rmse = []

ks = range(2,30)

 

for k in ks:

 

    knnr = KNeighborsRegressor(n_neighbors=k)

   

    X_train = df_train[explicativas].values

    y_train = df_train['idhm'].values.reshape(-1, 1)

    knnr.fit(X_train,y_train)

   

    X_test = df_test[explicativas].values

    y_test = df_test['idhm'].values.reshape(-1, 1)

    rmse.append(mean_squared_error(knnr.predict(X_test),y_test))
```

```python
# Plotando resultados num gráfico

 

plt.figure(figsize=(12,8))

 

plt.plot(rmse,marker='s',color='red')

 

# Opções de formatação

plt.ylabel('RMSE',fontsize=20)

plt.xlabel('Número de vizinhos (k)',fontsize=20)

plt.tick_params(axis='both', which='major', labelsize=12)

 

plt.xticks(range(len(rmse)),ks)

 

plt.grid()

 

plt.show()
```
