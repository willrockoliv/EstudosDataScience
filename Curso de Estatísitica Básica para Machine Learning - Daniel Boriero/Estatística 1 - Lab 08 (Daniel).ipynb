```
#Crie três vetores ordenados com 500 números aleatórios, o primeiro distribuído gaussianamente, o segundo log-gaussianamente e o terceiro uma Pareto
# 1. o primeiro com centro em 36 e desvio padrão de 4, o qual suporemos que é uma distribuição de idade
# 2. o segundo com centro em 7000 e coeficiente de forma de 1000, o qual suporemos que é uma distribuição de renda
# 3. o terceiro com início em 0 e coeficiente de forma de 10000, o qual suporemos que é uma distribuição de riqueza

# dicas: - pode utilizar os métodos rvs das funções na biblioteca stat da scipy
#        - https://docs.scipy.org/doc/scipy/reference/stats.html

import scipy.stats as st

exemplo = st.norm.rvs(loc=1000, scale=10, size=100, random_state=None)
```

```
import scipy.stats as st

idade = st.norm.rvs(loc=36, scale=4, size=500, random_state=None)
renda = st.lognorm.rvs(s=2, loc=4000, scale=1, size=500, random_state=None)
riqueza = st.pareto.rvs(b=100000, loc=0, scale=50000, size=500, random_state=None)
```

```
#Plote o histograma das duas distribições, e em cima do gráfico a média, desvio padrão, obliquidade de Pearson e curtosis
# dicas: - pode utilizar o parâmetro "log=True" para mostrar um gráfico em escala log que seja muito assimétrico
#        - utilize as funções "Summary statistics" da biblioteca stats

import matplotlib.pyplot as plt
import numpy as np

f, (ax1,ax2,ax3) = plt.subplots(3,1,figsize=(8, 8))

n, l, m, v, s, k = st.describe(exemplo)
ax1.hist(exemplo,density=True)
ax1.set_title('média='+str(np.round(m,2))+', desvio='+str(np.round(np.sqrt(v),2))+', g_p='+str(np.round(s,2))+', curtosis='+str(np.round(k,2)))

n, l, m, v, s, k = st.describe(exemplo)
ax2.hist(exemplo,density=True)
ax2.set_title('média='+str(np.round(m,2))+', desvio='+str(np.round(np.sqrt(v),2))+', g_p='+str(np.round(s,2))+', curtosis='+str(np.round(k,2)))


n, l, m, v, s, k = st.describe(exemplo)
ax3.hist(exemplo,density=True)
ax3.set_title('média='+str(np.round(m,2))+', desvio='+str(np.round(np.sqrt(v),2))+', g_p='+str(np.round(s,2))+', curtosis='+str(np.round(k,2)))


f.tight_layout()
f.subplots_adjust(hspace=1)
display(f)
```

```
import matplotlib.pyplot as plt
import numpy as np

f, (ax1,ax2,ax3) = plt.subplots(3,1,figsize=(8, 8))

n, l, m, v, s, k = st.describe(idade)
ax1.hist(idade,density=True)
ax1.set_title('média='+str(np.round(m,2))+', desvio padrão='+str(np.round(np.sqrt(v),2))+', obliquidade_pearson='+str(np.round(s,2))+', curtosis='+str(np.round(k,2)))

n, l, m, v, s, k = st.describe(renda)
ax2.hist(renda,density=True)
ax2.set_title('média='+str(np.round(m,2))+', desvio padrão='+str(np.round(np.sqrt(v),2))+', obliquidade_pearson='+str(np.round(s,2))+', curtosis='+str(np.round(k,2)))

n, l, m, v, s, k = st.describe(riqueza)
ax3.hist(riqueza,density=True)
ax3.set_title('média='+str(np.round(m,2))+', desvio padrão='+str(np.round(np.sqrt(v),2))+', obliquidade_pearson='+str(np.round(s,2))+', curtosis='+str(np.round(k,2)))

f.tight_layout()
f.subplots_adjust(hspace=1)
display(f)
```
