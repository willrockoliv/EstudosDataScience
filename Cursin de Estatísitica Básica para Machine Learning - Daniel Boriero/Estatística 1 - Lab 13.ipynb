```
#Utilize as duas distribuições normais abaixo, a primeira com 3666 e a segunda com 611 elementos, com centros em 4000 e 4100, e desvios padrões de 1.500 e 1000, respectivamente.

#Esses dois grupos representam a distribuição de renda de dois grupos que desconfiamos terem um perfil de renda diferente. Utilize teste de hipótese para provar se isso é verdade.

# Passos: 1. Defina sua hipótese nula
#         2. Faça simulações da sua hipótese nula
#         3. Verifique qual a proporção das simulações em que ocorre uma diferença tão grande quanto a da observada originalmente

#dicas: -https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html
#        norm.rvs(loc=0, scale=1, size=1, random_state=None) ##Random variates.


#       -https://docs.python.org/3/library/random.html       
#       -random.shuffle(vetor) 
```

```
import scipy.stats as st
renda1 = st.norm.rvs(loc=4000, scale=1500, size=3666, random_state=None)
renda2 = st.norm.rvs(loc=4100, scale=1000, size=611, random_state=None)
```

```
import matplotlib.pyplot as plt

f, ax1 = plt.subplots(figsize=(6,4))

ax1.hist(renda1, label='android', density=True, histtype='step')
ax1.hist(renda2, label='ios', density=True, histtype='step')
ax1.set_xlabel('renda')
ax1.set_ylabel('frequência')
ax1.legend()

f.tight_layout()
display(f)
```

```
# 1. Defina sua hipótese nula
# A hipóotese nula é que são as mesmas
h0 = list(renda1) + list(renda2)
diff_abs_orig = np.abs(np.mean(renda1) - np.mean(renda2))
diff_abs_orig
```

```
# 2. Faça simulações da sua hipótese nula

import numpy as np
import random as rd

n = 1000
diff_abs_sim_vec = []
for i in range(n):
  rd.shuffle(h0)
  renda1_sim = h0[:len(renda1)]
  renda2_sim = h0[:len(renda2)]

  diff_abs_sim = np.abs(np.mean(renda1_sim) - np.mean(renda2_sim))
  diff_abs_sim_vec.append(diff_abs_sim)
  
diff_abs_sim_vec
```

```
# 3. Verifique qual a proporção das simulações em que ocorre uma diferença tão grande quanto a da observada originalmente

len([x for x in diff_abs_sim_vec if x>diff_abs_orig]) / len(diff_abs_sim_vec)
```
