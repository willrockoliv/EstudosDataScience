```
#Crie e plote uma distribuição analítica que permita prever qual a probabilidade de um determinado evento ocorrer n vezes no próximo ano (desde não ocorrer a ocorrer dez vezes), 
#sendo que normalmente esse evento ocorre três vezes por ano

# dicas: - https://docs.scipy.org/doc/scipy/reference/stats.html
#        - utilize o método "pmf" para criar a distribuição analítica discreta
#        - pode usar a função linspace para criar um vetor de números distribuídos linearmente e funcionar como a ordenada (eixo x) da sua distribuição analítica

import numpy as np

x = np.linspace(0,10,10)

import matplotlib.pyplot as plt
f, ax1 = plt.subplots(figsize=(5, 4))

ax1.plot(x,[i**2 for i in x])

f.tight_layout()
display(f)
```

```
#Crie e plote uma distribuição analítica que permita prever qual a probabilidade de um sequência de sucessos em um jogo de cartas em 20 lances, sendo que um teste de Bernoulli foi constatado que a probabilidade é de 27%

# dicas: - https://docs.scipy.org/doc/scipy/reference/stats.html
#        - utilize o método "pmf" para criar a distribuição analítica discreta
#        - pode usar a função linspace para criar um vetor de números distribuídos linearmente e funcionar como a ordenada (eixo x) da sua distribuição analítica

import numpy as np
x = np.linspace(1,20,20)
```

```
import scipy.stats as st
distr_binomial = st.binom.pmf(x,20,0.27)
```

```
import numpy as np
import matplotlib.pyplot as plt

f, ax1 = plt.subplots(figsize=(5, 4))

ax1.plot(x,distr_binomial, linestyle='', marker='.')

f.tight_layout()
display(f)
```

```
#Qual o intervalo de tentativas tal que contenha 40% da população mais provável de ter sucesso? 
#dica:   - use o método "interval" da função "binom"
import scipy.stats as st
st.binom.interval(0.4,20,0.27) # R.: Quero ter 40% do máximo de propbabilidades. Então jogue de 4 a 6 vezes.
```

```
#Plote novamente o gráfico acima e coloque as linhas verticais correspondentes à esse intervalo
import numpy as np
import matplotlib.pyplot as plt

f, ax1 = plt.subplots(figsize=(5, 4))

ax1.plot(x,distr_binomial, linestyle='', marker='.')
ax1.axvline(4,label='40%',color='red')
ax1.axvline(6,label='60%',color='red')

ax1.legend()
f.tight_layout()
display(f)
```
