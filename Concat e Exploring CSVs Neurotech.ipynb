```python
import pandas as pd
```

```python
def path(i):
    return "C:/GB/GBConnect/Extrações VARS/Neurotech/01 retorno/Processados-20200730T145806Z-001/Processados/backtest_entrega_GuiaBolso_03072020_{}.csv".format(i)
```

```python
df_list = []
```

```python
def import_csvs():
    for i in range(0,33):
        if i < 10:
            path_csv = path('0'+str(i))
        else:
            path_csv = path(i)

        print(path_csv)
        df = pd.read_csv(path_csv)
        df["tag_origin"] = i
        df_list.append(df)
    
    return pd.concat(df_list)
```

```python
df_result = import_csvs()
```

```python
print(df_result.count())
print(df_result["user_id"].drop_duplicates().count())
print(df_result[" cpf"].drop_duplicates().count())
print(df_result[" reference_date"].drop_duplicates().count())
```

```python
df_result.to_csv("backtest_entrega_GuiaBolso_03072020_ALL.csv")
```
