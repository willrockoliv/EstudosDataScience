```python
import numpy as np
import pandas as pd
from sklearn.cluster import MeanShift, KMeans
from sklearn.datasets.samples_generator import make_blobs
import matplotlib.pyplot as plt
```

```python
centers = [[1,1], [5,5], [3,10]]
X, _ = make_blobs(n_samples=100, centers=centers, cluster_std=2, random_state=0)
```

```python
X[:5]
```

```python
df = pd.DataFrame(data=X, columns=["x", "y"])
df["cluster"] = _
df.head()
```

```python
plt.scatter(x=df["x"], y=df["y"], c=df["cluster"])
```

```python
ms = MeanShift(cluster_all=True, bandwidth=3)
ms.fit(X)
```

```python
ms.cluster_centers_
```

```python
df["cluster_mean_shift"] = ms.labels_
```

```python
plt.scatter(x=df["x"], y=df["y"], c=df["cluster_mean_shift"])
```
