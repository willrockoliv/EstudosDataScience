```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.datasets import load_iris

%matplotlib inline
```

```python
data = load_iris()
df = pd.DataFrame(data["data"],
                  columns=data["feature_names"]
                 )
df["target"] = data["target"]
df["target_names"] = (
    np.where(
        df["target"] == 0,
        data["target_names"][0],
        np.where(
            df["target"] == 1,
            data["target_names"][1],
            np.where(
                df["target"] == 2,
                data["target_names"][2],
                None
            )
        )
    )
)
```

```python
df
```

```python
sb.pairplot(df, hue="target")
```

```python
X = np.array(df.drop("target", 1).drop("target_names", 1))
X
```

### K-means

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3, random_state=0)
```

```python
kmeans.fit(X)
```

```python
kmeans.labels_
```

```python
df["k_classes"] = kmeans.labels_
```

```python
df
```

```python
sb.pairplot(df, hue="target")
```

```python
sb.pairplot(df, hue="k_classes")
```
