```python
df_agg2 = (
    df
    .groupby(["vehicleType", "gearbox"])
    .agg({"price":"mean"})
    .rename(columns={"price":"mean_price"})
    .reset_index()
)

f = plt.figure(figsize=(10,7))
sns.barplot(x="vehicleType", y='mean_price', hue="gearbox", data=df_agg2, palette='summer_r')
plt.title("Preço Médio por Tipo de Veículo Caixa de Câmbio", fontsize=20)
plt.xlabel("Tipo de Veículo", fontsize=20)
plt.ylabel("Preço Médio", fontsize=20)
```
