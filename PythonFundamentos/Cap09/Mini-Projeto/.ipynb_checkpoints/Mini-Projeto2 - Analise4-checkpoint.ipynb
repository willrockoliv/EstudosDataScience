```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style="white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding="latin-1")
```

```python
df.columnsumns
```

```python
# Calcule a média de preço por marca e por veículo
```

```python
df_agg1 = (
    df
    .groupby(["brand", "vehicleType"])
    .agg({"price":"mean"})
    .rename(columns={"price":"avgPrice"})
    .reset_index()
)
df_agg1[(df_agg1["brand"]=="volkswagen") & (df_agg1["vehicleType"].isin(["Other","coupe","suv","kleinwagen","limousine",]))]
```
