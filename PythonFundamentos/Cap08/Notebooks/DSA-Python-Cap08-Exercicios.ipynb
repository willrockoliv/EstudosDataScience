```python
# Crie um banco de dados no SQLite, crie uma tabela, insira registros, 
# consulte a tabela e retorne os dados em dataframe do Pandas

import sqlite3

def execute(command):
    # Criando uma conexão
    conn = sqlite3.connect('exercicio10.db')   
    # Criando um cursor
    cursor = conn.cursor()
    
    cursor.execute(command)
    fetchall = cursor.fetchall()
    
    conn.commit()
    cursor.close()
    conn.close()
    
    return fetchall
    
# Função para criar uma tabela
def create_table(table='', cols=[]):
    command = "CREATE TABLE IF NOT EXISTS {} (id INTEGER PRIMARY KEY AUTOINCREMENT NOT NULL, {})"
    command = command.format(table, ", ".join(cols))
    execute(command)
    
# Função para criar uma tabela
def drop_table(table):
    command = "DROP TABLE {}".format(table)
    execute(command)
    
# Função para inserir uma linha
def data_insert(table, cols=[], values=[]):
    command = "INSERT INTO {}({}) VALUES({})"
    command = command.format(table,", ".join(cols), ", ".join([str(i) if type(i) in (int,float) else "\'{}\'".format(i) for i in values]))
    execute(command)

def data_delete(table, _id=[]):
    command = "DELETE FROM {} WHERE id IN ({})"
    command = command.format(table, ", ".join([str(i) for i in _id]))
    execute(command)
    
# Leitura de dados
def data_select_all(table):
    fetchall = execute("SELECT * FROM {}".format(table))
    return [linha for linha in fetchall]
        
# Leitura de dados
def data_select_by_id(table, _id):
    fetchall = execute("SELECT * FROM {} WHERE id IN ({})".format(table, ",".join([str(i) for i in _id])))
    return [linha for linha in fetchall]
```

```python
table = 'produto'
```

```python
create_table(table, ['nome TEXT','valor REAL','qtd INTEGER'])
```

```python
data_insert(table, ['nome','valor','qtd'], ['notebook',5000.00,2])
```

```python
data_select_all(table)
```

```python
data_select_by_id(table, [2,3])
```

```python
data_delete(table, [1,2])
```

```python
data_select_all(table)
```

```python
import pandas as pd
df = pd.DataFrame(data_select_all(table), columns=['id','nome','valor','qtd'])
df
```
