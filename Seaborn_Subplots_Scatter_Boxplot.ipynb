```python
# Imports
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1 = pd.DataFrame(np.random.randint(0,100,100), columns=['a'])
df2 = pd.DataFrame(np.random.randint(100,200,100), columns=['a'])
df3 = pd.DataFrame(np.random.randint(0,1000,100), columns=['a'])

df = pd.concat([df1, df2, df3])

limit_sup = df['a'].quantile(0.95)
limit_inf = df['a'].quantile(0.05)

print(limit_inf)
print(limit_sup)
```

```python
nrows = 2
ncols = 1

plt.figure(figsize=(10,10))

plt.subplot(nrows, ncols, 1)
sns.scatterplot(data=df)

plt.subplot(nrows, ncols, 2)
sns.boxplot(data=df)
```
